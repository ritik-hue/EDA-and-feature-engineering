```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train=pd.read_csv('/content/drive/MyDrive/train.csv')
df_train.head()
df_test=pd.read_csv('/content/drive/MyDrive/test.csv')
df=df_train.append(df_test)
df.head()
df.drop(['User_ID'],axis=1,inplace=True)
```

axis=1 in drop means removing the value columnwise whereas in row it is axis=0

```python
df['Gender']=df['Gender'].map({'F':0,'M':1})
##df['Gender']=pd.get_dummies(df['Gender'],drop_first=True)
df.head()
```

or we can use pd.get_dummies to convert categorical
df['gender']=pd.get_dummies(df['gender'],drop_first=1)


```python
df['Age'].unique()
```

```python
##df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-55':5,'55+':6})

# Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
df['Age']= label_encoder.fit_transform(df['Age'])

##df['Age'].unique()
df.head()
```

for assigning values to age we can use map function or label encoding

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.drop('City_Category',inplace=True,axis=1)
```

```python
df.isnull().sum()
```

```python
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
sns.barplot(x='Product_Category_2',y='Purchase',hue='Gender',data=df)
```
