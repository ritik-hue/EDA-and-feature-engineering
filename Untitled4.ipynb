```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train=pd.read_excel('/content/drive/MyDrive/Data_Train.xlsx')
df_train.head()
```

```python
df_test=pd.read_excel('/content/drive/MyDrive/Test_set.xlsx')
final_df=df_train.append(df_test)
final_df.head()
```

```python
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_minute']=final_df['Dep_Time'].str.split(':').str[1]
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
final_df['Dep_minute']=final_df['Dep_minute'].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
```
